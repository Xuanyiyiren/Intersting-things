回溯算法伪代码框架

```
function Backtrack(PartialSolution):
    if PartialSolution is a complete solution:
        report PartialSolution as a valid solution
        return
    for each choice in RemainingChoices:
        if choice is feasible:
            make the choice
            Backtrack(PartialSolution with choice)
            undo the choice (backtrack)
```

````python
def solve_n_queens(n):
    def is_safe(board, row, col):
        # 检查同一列是否有皇后
        for i in range(row):
            if board[i][col] == 1:
                return False
        # 检查左上角对角线是否有皇后
        for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        # 检查右上角对角线是否有皇后
        for i, j in zip(range(row, -1, -1), range(col, n)):
            if board[i][j] == 1:
                return False
        return True

    def backtrack(board, row):
        if row == n:
            # 如果所有行都放置了皇后，找到一个解
            solutions.append([row[:] for row in board])
            return
        for col in range(n):
            if is_safe(board, row, col):
                # 在当前位置放置皇后
                board[row][col] = 1
                # 递归到下一行
                backtrack(board, row + 1)
                # 回溯，移除当前位置的皇后
                board[row][col] = 0

    solutions = []
    board = [[0]*n for _ in range(n)]
    backtrack(board, 0)
    return solutions
````

````python
# 示例：求解4皇后问题
solutions = solve_n_queens(4)
for solution in solutions:
    for row in solution:
        print(row)
    print()
````
