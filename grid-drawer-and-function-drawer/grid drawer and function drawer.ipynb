```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
plt.figure(figsize=(8, 8))
x = np.linspace(-10, 10, 1000)
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
eps = 1e-1
plt.xlim(-10-eps,10+eps)
plt.ylim(-10-eps,10+eps)
# set x,y axis at (0,0)
plt.xticks(np.arange(-10, 10+1, 1))
plt.yticks(np.arange(-10, 10+1, 1))
plt.grid()
# tight layout
plt.tight_layout()
plt.savefig('grid.pdf')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
x = np.linspace(-10, 10, 1000)
plt.plot(x, np.sin(x), 'b', label=r'$\sin x$')
plt.plot(x, np.cos(x), 'r', label=r'$\cos x$')
plt.plot(x, np.exp(x), 'g', label=r'$\mathrm{e}^x$')
plt.plot(x, x**2 , 'orange', label=r'$x^2$')
plt.plot(x, x, 'k--')

ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
eps = 1e-1
plt.xlim(-10-eps,10+eps)
plt.ylim(-10-eps,10+eps)
# set x,y axis at (0,0)
plt.xticks(np.arange(-10, 10+1, 1))
plt.yticks(np.arange(-10, 10+1, 1))
plt.legend()
plt.grid()
# tight layout
plt.tight_layout()
plt.savefig('function_plot.pdf')
plt.show()
```
