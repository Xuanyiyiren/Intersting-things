# PDF to SVG

Sometimes we a picture with `.pdf` is not accepted, but `.svg` can be groundly received, so I have to convert pdf to svg. But another time is that some free converter cannot display the fonts well. However, the `fitz` python package does it well.

```python
def pdf2svg(pdf_file,svg_file,page = 0):
    '''
    require fitz package
    pdf_file: the path of your pdf file to be converted.
    svg_file: the path to store the output svg file.
    page: which page to be converted.
    '''
    import fitz
    import os
    pdf_extension = os.path.splitext(pdf_file)[1]
    if pdf_extension == '':
        pdf_file += '.pdf'
    elif pdf_extension == '.pdf':
        pass
    else:
        raise TypeError("Not pdf!")

    svg_extension = os.path.splitext(svg_file)[1]
    if svg_extension == '':
        svg_file += '.svg'

    doc = fitz.open(pdf_file)
    page = doc.load_page(page)
    svg = page.get_svg_image()
    with open(svg_file, 'w') as f:
        f.write(svg)
```

```python
pdf2svg('periodic_table.pdf','periodic_table')
```

```python
pdf2svg('SM_isospin','SM_isospin',page = 1)
```

```python
pdf2svg('Bhabha1.pdf','Bhabha1')
pdf2svg('Bhabha2.pdf','Bhabha2')
pdf2svg('emu_emu.pdf','emu_emu')
```
