```python
import matplotlib.pyplot as plt
import numpy as np
import cmath as cm
import sympy as sy
# cm.cos(cm.pi)
def visualize_complex_numbers(complex_numbers, point_color='b'):
    complex_numbers=np.array(complex_numbers)
    # 提取实部和虚部
    real_parts = complex_numbers.real
    imaginary_parts = complex_numbers.imag

    # 创建一个复数平面图
    plt.figure(figsize=(8, 6))
    plt.scatter(real_parts, imaginary_parts, c=point_color, marker='o')
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.title('Complex Number Visualization')
    plt.grid(True)
    plt.xlim(-1.2, 1.2)
    plt.ylim(-1.2, 1.2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(False)
    # plt.lim(-2, 2)
    plt.show()
```

```python
equ=0
grp=[cm.exp(i*1j*cm.pi/6) for i in range(1,13)]
visualize_complex_numbers(grp)
grp
```

```python
def generate_gruop(gen):
    grp=[gen]
    base=gen*gen
    maxgeneration=100
    while abs(base-gen)>1e-6:
        grp.append(base)
        base=base*gen
        maxgeneration-=1
        if maxgeneration==0:
            print('达到生成数上限，该元素的阶至少为100')
            break
    return np.array(grp)

a=generate_gruop(cm.exp(1j*cm.pi/6))
visualize_complex_numbers(a)
# generate_gruop(cm.exp(1j*cm.pi**2))
```

```python
for i,k in enumerate(grp):
    print(f"我们生成了第{i+1}个群元{k}的生成群")
    visualize_complex_numbers(generate_gruop(k))
```
